```python
# Function to deposit money
def deposit(amount):
    global balance
    balance = balance + amount
    print("Amount Deposited:", amount)
    print("Updated Balance:", balance)
```

```python
 #Function to withdraw money
def withdraw(amount):
    global balance
    if amount <= balance:
        balance = balance - amount
        print("Amount Withdrawn:", amount)
        print("Remaining Balance:", balance)
    else:
        print("Insufficient Balance!")
```

```python
# Function to check balance
def check_balance():
    print("Current Balance:", balance)
```

```python
# Menu-driven program
while True:
    print("\n--- Bank Account Manager ---")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        amount = float(input("Enter amount to deposit: "))
        deposit(amount)

    elif choice == 2:
        amount = float(input("Enter amount to withdraw: "))
        withdraw(amount)

    elif choice == 3:
        check_balance()

    elif choice == 4:
        print("Thank you for using the Bank Manager")
        break

    else:
        print("Invalid Choice")
```

#2. Weather Data Analyzer

#a. Analyze weather data (temperature, humidity, etc.) for a given region
#b. Use variables and data types to store weather data
#c. Implement conditional statements to determine weather conditions
#(sunny, rainy, etc.)
#d. Use functions to calculate average temperature and humidity, and print results

```python
# Weather Data
region = "Kerala"
temperature = [30, 32, 31, 29, 33]
humidity = [70, 65, 72, 68, 75]
```

```python
# Function to calculate average temperature
def calculate_avg_temp(temp_list):
    return sum(temp_list) / len(temp_list)
```

```python
# Function to calculate average humidity
def calculate_avg_humidity(hum_list):
    return sum(hum_list) / len(hum_list)
```

```python
# Function to determine weather condition
def weather_condition(avg_temp, avg_humidity):
    if avg_temp > 35:
        print("Weather Condition: Very Hot")
    elif avg_temp > 30:
        print("Weather Condition: Sunny")
    elif avg_humidity > 70:
        print("Weather Condition: Rainy or Humid")
    else:
        print("Weather Condition: Pleasant")
```

```python
# Calculate averages
avg_temp = calculate_avg_temp(temperature)
avg_humidity = calculate_avg_humidity(humidity)
```

```python
# Display results
print("Region:", region)
print("Average Temperature:", avg_temp)
print("Average Humidity:", avg_humidity)

weather_condition(avg_temp, avg_humidity)
```
